```python
# ═══════════════════════════════════════════════════════════════════════════
# InSDN TCN Binary Classification  –  Google Colab (T4 GPU)
# Sections: 1-Imports | 2-Upload/Load | 3-EDA | 4-Clean | 5-Scale/PCA
#           6-Split   | 7-TCN Model   | 8-Train | 9-Evaluate
# ═══════════════════════════════════════════════════════════════════════════

# ── SECTION 1 – Install & Import ─────────────────────────────────────────
# Uncomment if running in Colab for the first time:
# !pip install -q pandas numpy scikit-learn matplotlib seaborn tensorflow

import os, warnings, random, time
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve, ConfusionMatrixDisplay, f1_score
)
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks, regularizers

SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("TensorFlow :", tf.__version__)
print("Pandas     :", pd.__version__)
print("NumPy      :", np.__version__)

gpus = tf.config.list_physical_devices("GPU")
if gpus:
    print(f"\n✅  GPU detected: {gpus[0].name}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("\n⚠️  No GPU – running on CPU")

# ── SECTION 2 – Upload & Load consolidated CSV ───────────────────────────
from google.colab import files

CSV_FILENAME = "insdn_consolidated.csv"

if os.path.exists(CSV_FILENAME):
    print(f"✅  Found existing file: {CSV_FILENAME} – skipping upload.")
    CSV_PATH = CSV_FILENAME
else:
    print(f"'{CSV_FILENAME}' not found. Please upload it …")
    uploaded = files.upload()
    CSV_PATH = list(uploaded.keys())[0]
    print(f"✅  Uploaded: {CSV_PATH}")

df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()
print(f"\nShape        : {df.shape}")
print(f"Memory usage : {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
display(df.head(3))

# ── SECTION 3 – EDA ──────────────────────────────────────────────────────
feature_cols = [c for c in df.columns if c != "Label"]
print(f"\nFeatures : {len(feature_cols)}")
print(f"Samples  : {len(df):,}")
print(f"\nLabel distribution:\n{df['Label'].value_counts()}")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
counts = df["Label"].value_counts()
axes[0].bar(["Benign (0)", "Attack (1)"], counts.values,
            color=["steelblue", "tomato"], edgecolor="black")
axes[0].set_title("Class Distribution")
axes[0].set_ylabel("Samples")
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 500, f"{v:,}", ha="center", fontweight="bold")
axes[1].pie(counts.values, labels=["Benign", "Attack"],
            autopct="%1.1f%%", colors=["steelblue", "tomato"],
            startangle=90, wedgeprops=dict(edgecolor="white"))
axes[1].set_title("Class Proportion")
plt.tight_layout()
plt.savefig("eda_class_distribution.png", dpi=150, bbox_inches="tight")
plt.show()

missing = df[feature_cols].isnull().sum()
print(f"Columns with NaN : {(missing > 0).sum()}")
print(f"Total NaN cells  : {missing.sum()}")

# ── SECTION 4 – Safety-net Cleaning ──────────────────────────────────────
ID_COLS = ["Flow ID", "Src IP", "Src Port", "Dst IP", "Dst Port", "Timestamp"]
existing_ids = [c for c in ID_COLS if c in df.columns]
if existing_ids:
    df.drop(columns=existing_ids, inplace=True)
    print(f"Dropped: {existing_ids}")

feature_cols = [c for c in df.columns if c != "Label"]
inf_before   = np.isinf(df[feature_cols].select_dtypes(include=np.number)).sum().sum()
df[feature_cols] = df[feature_cols].replace([np.inf, -np.inf], np.nan)
nan_before = df.isnull().sum().sum()
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

print(f"Inf replaced : {inf_before:,}")
print(f"NaN removed  : {nan_before:,}")
print(f"Final shape  : {df.shape}")
print(df["Label"].value_counts())

# ── SECTION 5 – Configuration, Scaling & PCA ─────────────────────────────
USE_PCA       = True    # Set False to skip PCA and use all scaled features
PCA_VARIANCE  = 0.95    # Fraction of variance to retain
TEST_SIZE     = 0.20
VAL_SIZE      = 0.10
BATCH_SIZE    = 2048
EPOCHS        = 100
LEARNING_RATE = 1e-3

feature_cols = [c for c in df.columns if c != "Label"]
X = df[feature_cols].values.astype(np.float32)
y = df["Label"].values.astype(np.int32)

scaler   = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"\nAfter StandardScaler – mean≈{X_scaled.mean():.4f}, std≈{X_scaled.std():.4f}")

if USE_PCA:
    pca     = PCA(n_components=PCA_VARIANCE, random_state=SEED)
    X_final = pca.fit_transform(X_scaled).astype(np.float32)
    print(f"PCA: {X_scaled.shape[1]} → {pca.n_components_} components "
          f"({pca.explained_variance_ratio_.sum()*100:.2f}% variance retained)")
    fig, ax = plt.subplots(figsize=(9, 4))
    cumvar = np.cumsum(pca.explained_variance_ratio_) * 100
    ax.plot(range(1, len(cumvar) + 1), cumvar, marker="o", markersize=4, color="steelblue")
    ax.axhline(PCA_VARIANCE * 100, color="tomato", linestyle="--",
               label=f"{PCA_VARIANCE*100:.0f}% threshold")
    ax.set_xlabel("Components"); ax.set_ylabel("Cumulative Variance (%)")
    ax.set_title("PCA – Cumulative Explained Variance")
    ax.legend(); ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig("pca_explained_variance.png", dpi=150, bbox_inches="tight")
    plt.show()
else:
    print("PCA skipped – using all scaled features")
    X_final = X_scaled

N_FEATURES = X_final.shape[1]
print(f"Final feature count : {N_FEATURES}")

# ── SECTION 6 – Train / Val / Test Split & Reshape ───────────────────────
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X_final, y, test_size=TEST_SIZE, random_state=SEED, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val,
    test_size=VAL_SIZE / (1 - TEST_SIZE),
    random_state=SEED, stratify=y_train_val)

print(f"Train : {X_train.shape}  Attack={y_train.sum():,}  Benign={(y_train==0).sum():,}")
print(f"Val   : {X_val.shape}   Attack={y_val.sum():,}  Benign={(y_val==0).sum():,}")
print(f"Test  : {X_test.shape}   Attack={y_test.sum():,}  Benign={(y_test==0).sum():,}")

classes       = np.unique(y_train)
class_w_arr   = compute_class_weight("balanced", classes=classes, y=y_train)
class_weights = dict(zip(classes.tolist(), class_w_arr.tolist()))
print(f"Class weights: {class_weights}")

# TCN input shape: (N, N_FEATURES, 1) – each feature is a time-step with 1 channel
X_train_3d = X_train.reshape(-1, N_FEATURES, 1)
X_val_3d   = X_val.reshape(-1, N_FEATURES, 1)
X_test_3d  = X_test.reshape(-1, N_FEATURES, 1)
print(f"3-D shapes – Train:{X_train_3d.shape}  Val:{X_val_3d.shape}  Test:{X_test_3d.shape}")

# ── SECTION 7 – Custom TCN Architecture ──────────────────────────────────
def tcn_residual_block(x, filters, kernel_size, dilation_rate, dropout_rate=0.2):
    """Two dilated causal Conv1D layers + BN + ReLU + SpatialDropout + residual skip."""
    residual = x
    for _ in range(2):
        x = layers.Conv1D(
            filters=filters, kernel_size=kernel_size,
            dilation_rate=dilation_rate, padding="causal",
            kernel_initializer="he_normal",
            kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.SpatialDropout1D(dropout_rate)(x)
    if residual.shape[-1] != filters:
        residual = layers.Conv1D(filters, kernel_size=1, padding="same")(residual)
    return layers.Add()([x, residual])

def build_tcn(input_shape, num_filters=64, kernel_size=3,
              dilation_rates=None, dropout_rate=0.2, dense_units=None):
    if dilation_rates is None:
        dilation_rates = [1, 2, 4, 8, 16, 32]
    if dense_units is None:
        dense_units = [128, 64]
    inputs = keras.Input(shape=input_shape, name="input")
    x = inputs
    for d in dilation_rates:
        x = tcn_residual_block(x, num_filters, kernel_size, d, dropout_rate)
    x = layers.GlobalAveragePooling1D()(x)
    for units in dense_units:
        x = layers.Dense(units, kernel_initializer="he_normal",
                         kernel_regularizer=regularizers.l2(1e-4))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)
    outputs = layers.Dense(1, activation="sigmoid", name="output")(x)
    return keras.Model(inputs, outputs, name="TCN_InSDN")

model = build_tcn(
    input_shape=(N_FEATURES, 1),
    num_filters=64,
    kernel_size=3,
    dilation_rates=[1, 2, 4, 8, 16, 32],
    dropout_rate=0.2,
    dense_units=[128, 64],
)
model.summary()
print(f"\nTotal trainable parameters: {model.count_params():,}")

# ── SECTION 8 – Compile & Train ───────────────────────────────────────────
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        keras.metrics.AUC(name="auc"),
        keras.metrics.Precision(name="precision"),
        keras.metrics.Recall(name="recall"),
    ],
)

CKPT_PATH = "best_tcn_insdn.keras"

print("=" * 60)
print(f"Training on {len(X_train_3d):,} samples | val on {len(X_val_3d):,}")
print(f"Batch size : {BATCH_SIZE}  |  Max epochs : {EPOCHS}")
print(f"GPU        : {gpus[0].name if gpus else 'CPU'}")
print("=" * 60)

t0 = time.time()
history = model.fit(
    X_train_3d, y_train,
    validation_data=(X_val_3d, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    class_weight=class_weights,
    callbacks=[
        callbacks.EarlyStopping(monitor="val_auc", patience=15,
                                restore_best_weights=True, mode="max", verbose=1),
        callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                                    patience=7, min_lr=1e-6, verbose=1),
        callbacks.ModelCheckpoint(filepath=CKPT_PATH, monitor="val_auc",
                                  save_best_only=True, mode="max", verbose=1),
        callbacks.CSVLogger("training_log.csv", append=False),
    ],
    verbose=1,
)
print(f"\nTraining finished in {(time.time()-t0)/60:.1f} min")
print(f"Best model saved to: {CKPT_PATH}")

hist_df = pd.read_csv("training_log.csv")
print("\nTraining log (last 5 epochs):")
print(hist_df.tail(5).to_string(index=False))

# ── SECTION 9 – Evaluate on Test Set ─────────────────────────────────────
best_model   = keras.models.load_model(CKPT_PATH)
test_results = best_model.evaluate(X_test_3d, y_test, batch_size=BATCH_SIZE, verbose=1)

print("\n--- Keras metrics ---")
for name, val in zip(best_model.metrics_names, test_results):
    print(f"  {name:<15}: {val:.6f}")

y_prob = best_model.predict(X_test_3d, batch_size=BATCH_SIZE, verbose=0).flatten()
THRESHOLD = 0.5
y_pred    = (y_prob >= THRESHOLD).astype(int)

roc_auc = roc_auc_score(y_test, y_prob)
cm      = confusion_matrix(y_test, y_pred)

print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred, target_names=["Benign", "Attack"], digits=6))
print(f"ROC-AUC Score : {roc_auc:.6f}")

tn, fp, fn, tp = cm.ravel()
print("\n--- Security Metrics ---")
print(f"  TP (detected attacks) : {tp:,}")
print(f"  TN (correct benign)   : {tn:,}")
print(f"  FP (false alarms)     : {fp:,}")
print(f"  FN (missed attacks)   : {fn:,}")
print(f"  Detection Rate        : {tp/(tp+fn+1e-9):.4f}")
print(f"  Specificity           : {tn/(tn+fp+1e-9):.4f}")
print(f"  False Alarm Rate      : {fp/(fp+tn+1e-9):.4f}")

# ── Plot 1 – Loss ─────────────────────────────────────────────────────────
hist = history.history
ep   = range(1, len(hist["loss"]) + 1)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(ep, hist["loss"],     label="Train", color="steelblue")
ax.plot(ep, hist["val_loss"], label="Val",   color="tomato", linestyle="--")
ax.set_title("Loss"); ax.set_xlabel("Epoch"); ax.set_ylabel("Binary Cross-Entropy")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("plot_loss.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_loss.png")

# ── Plot 2 – Accuracy ─────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(ep, hist["accuracy"],     label="Train", color="steelblue")
ax.plot(ep, hist["val_accuracy"], label="Val",   color="tomato", linestyle="--")
ax.set_title("Accuracy"); ax.set_xlabel("Epoch"); ax.set_ylabel("Accuracy")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("plot_accuracy.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_accuracy.png")

# ── Plot 3 – AUC ──────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(ep, hist["auc"],     label="Train", color="steelblue")
ax.plot(ep, hist["val_auc"], label="Val",   color="tomato", linestyle="--")
ax.set_title("AUC"); ax.set_xlabel("Epoch"); ax.set_ylabel("AUC")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("plot_auc.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_auc.png")

# ── Plot 4 – Precision & Recall ───────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(ep, hist["precision"],     label="Train Precision", color="green")
ax.plot(ep, hist["val_precision"], label="Val Precision",   color="green",  linestyle="--")
ax.plot(ep, hist["recall"],        label="Train Recall",    color="purple")
ax.plot(ep, hist["val_recall"],    label="Val Recall",      color="purple", linestyle="--")
ax.set_title("Precision & Recall"); ax.set_xlabel("Epoch")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("plot_precision_recall.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_precision_recall.png")

# ── Plot 5 – Confusion Matrix ─────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(6, 5))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Benign", "Attack"])
disp.plot(ax=ax, colorbar=False, cmap="Blues",
          values_format=",d", xticks_rotation="horizontal")
ax.set_title(f"Confusion Matrix (threshold={THRESHOLD})")
plt.tight_layout()
plt.savefig("plot_confusion_matrix.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_confusion_matrix.png")

# ── Plot 6 – ROC Curve ────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 6))
fpr, tpr, _ = roc_curve(y_test, y_prob)
ax.plot(fpr, tpr, color="steelblue", lw=2, label=f"AUC = {roc_auc:.4f}")
ax.plot([0, 1], [0, 1], color="gray", linestyle="--", label="Random")
ax.set_xlabel("False Positive Rate"); ax.set_ylabel("True Positive Rate")
ax.set_title("ROC Curve"); ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("plot_roc_curve.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot_roc_curve.png")

print("\nAll plots saved individually:")
print("  plot_loss.png | plot_accuracy.png | plot_auc.png")
print("  plot_precision_recall.png | plot_confusion_matrix.png | plot_roc_curve.png")
print("Model & logs: best_tcn_insdn.keras | training_log.csv")
```
